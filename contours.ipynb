```python
from ocr.line_detector import find_contours
from PIL import Image

import cv2

template = cv2.imread("img_templates/left-top.png")
cnt_s = find_contours(template)
for i, cnt in enumerate(cnt_s):
    print(i, cv2.contourArea(cnt))
cv2.drawContours(template, [cnt_s[1]], -1, (0, 255, 0), 3)
Image.fromarray(template)
```
